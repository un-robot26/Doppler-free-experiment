```python
'''
Python code for calculations of relavant physical quantities.
Goal:
Calculate the values of coupling constants A and B and their uncertainties
'''
### physical constants, SI units ###
c = 299792458 # (m/s)
n_air = 1.0003

# length between F-P mirrors (m)
L = 0

```

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.signal import find_peaks
import numpy as np
# Path to your file
path1 = r'/Users/charlieglass/jlab2-spring26/exp1-doppler-free/Doppler-free-experiment/photodiode-CH1.CSV'
path2 = r'/Users/charlieglass/jlab2-spring26/exp1-doppler-free/Doppler-free-experiment/voltage-ramp-CH2.CSV'
# This CSV has metadata in cols 02, and the actual waveform in cols 3 (time) and 4 (voltage)
df1 = pd.read_csv(path1, header=None, usecols=[3, 4], names=["time_s", "voltage_V"])
df2 = pd.read_csv(path2, header=None, usecols=[3, 4], names=["time_s", "voltage_V"])

# Ensure numeric + drop any bad rows
df1["time_s"] = pd.to_numeric(df1["time_s"], errors="coerce")
df1["voltage_V"] = pd.to_numeric(df1["voltage_V"], errors="coerce")
df1 = df1.dropna()

df2["time_s"] = pd.to_numeric(df2["time_s"], errors="coerce")
df2["voltage_V"] = pd.to_numeric(df2["voltage_V"], errors="coerce")
df2 = df2.dropna()

# chop off data when the ramp goes up to fast
df1["time_s"] = df1["time_s"][:1801]
df1["voltage_V"] = df1["voltage_V"][:1801]

time = df1["time_s"]
signal = df1["voltage_V"] 

plt.figure()
plt.plot(df2["time_s"], df2["voltage_V"])
plt.xlabel("Time (s)")
plt.ylabel("Voltage (V)")
plt.title("CH2 Waveform")
plt.grid(True)
plt.tight_layout()
plt.show()

peaks, _ = find_peaks(signal, height=0.009, distance=75) 
peak_times = time[peaks]
plt.figure()
plt.plot(df1["time_s"].values[peaks], df1["voltage_V"].values[peaks], "rx", label="Detected Peaks")
plt.plot(df1["time_s"], df1["voltage_V"])
plt.xlabel("Time (s)")
plt.ylabel("Voltage (V)")
plt.title("CH1 Waveform")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# array of peak indices
peak_indices = np.arange(len(peaks))
# fit polynomial to index as function of time
coeffs = np.polyfit(peak_times, peak_indices, deg=2)
# convert to frequency using free spectral range and time polynomial
free_spectral_range = c/(2*L*n_air)
t_polynomial = coeffs[0] + time*coeffs[1] + (time**2)*coeffs[2]
frequency = free_spectral_range*t_polynomial
```

```python
# calculating coupling constants for hyperfine structure
# 5^2P_{3/2} state will be referred to as upper_state, 5S_{1/2} will be ground_state

# frequency between upper and ground state without hyperfine splitting
v_J = c/(n_air*780e-9)

# frequency measurement dictionary (real measurements to be filled in, just using theoretical ones for now)
frequencies = {'a':v_J -2793.2e6,'c':v_J -2793.2e6 + 157.2e6,'d':v_J+0,'g':v_J+4041.5e6,'i':v_J+0}

# list of quantum number changes in the real transitions
# a,c,d,g,i [(ground,upper)...]
# transitions = [(1,2,'a'),(1,1,'c'),(2,3,'d'),(2,2,'g'),(2,1,'i')]
transitions = [(2,1,'a'),(2,2,'c'),(2,3,'d'),(1,1,'g'),(1,2,'i')] # I question these F values for the transitions

# Nuclear spin, 3/2 for Rb 87 (28%) and 5/2 for Rb 85 (72%) 
I = 3/2

# ti and tj represent any transition, we just need two to calculate A and B
ti = transitions[0]
tj = transitions[1]
tk = transitions[3]
# i
fi_up = ti[1]
fi_gd = ti[0]
Ci_up = fi_up*(fi_up+1) - I*(I+1) - 3.75 
Ci_gd = fi_gd*(fi_gd+1) - I*(I+1) - .75 
alphai = (3*(fi_up*(fi_up+1)-3.75-I*(I+1))*((fi_up*(fi_up+1)-3.75-I*(I+1))+1) - 4*I*(I+1)*1.5*(1.5+1))/(2*I*(2*I-1)*1.5*(2*1.5-1)) 
# j
fj_up = tj[1]
fj_gd = tj[0]
Cj_up = fj_up*(fj_up+1) - I*(I+1) - 3.75 
Cj_gd = fj_gd*(fj_gd+1) - I*(I+1) - .75 
alphaj = (3*(fj_up*(fj_up+1)-3.75-I*(I+1))*((fj_up*(fj_up+1)-3.75-I*(I+1))+1) - 4*I*(I+1)*1.5*(1.5+1))/(2*I*(2*I-1)*1.5*(2*1.5-1)) 
# k
fk_up = tk[1]
fk_gd = tk[0]
Ck_up = fk_up*(fk_up+1) - I*(I+1) - 3.75 
Ck_gd = fk_gd*(fk_gd+1) - I*(I+1) - .75 
alphak = (3*(fk_up*(fk_up+1)-3.75-I*(I+1))*((fk_up*(fk_up+1)-3.75-I*(I+1))+1) - 4*I*(I+1)*1.5*(1.5+1))/(2*I*(2*I-1)*1.5*(2*1.5-1)) 

transition_matrix = np.array([[Ci_up/2, -Ci_gd/2, alphai/2],
                              [Cj_up/2, -Cj_gd/2, alphaj/2],
                              [Ck_up/2, -Ck_gd/2, alphak/2]])
frequency_vector = np.array([frequencies[ti[2]] - v_J, frequencies[tj[2]] - v_J, frequencies[tk[2]] - v_J])
inverse = np.linalg.inv(transition_matrix)
answer = np.dot(inverse, frequency_vector)                  

# coupling constants
A_p = answer[0]
A_s = answer[1]
B = answer[2]

print(A_p*1e-6, B*1e-6, A_s*1e-6) # lab guide shows A = 84.85 MHz, B = 12.51 MHz
```
